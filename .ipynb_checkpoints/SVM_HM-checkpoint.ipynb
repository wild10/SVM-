## SIMPLE SVM con SKlearn

Este es un simple ejemplo de SVM, con datos generados, aleatoriamente para este fin lo qu pretendemos hacer es el entrenamiento y la preuba de los diferentes kernels para nuestros datos y poder finalmente tener una clasificación y la precision de nuestro test.

### Generación y extraccion de data

para esto vamos hacer uso de un generador `make_blobs` de sklearn para luego trabajar en vase a esta data generada y probar el metodo SVM para clasificacion , entreno y test con los diferentes kernels, por lo tanto acontinuacion vamos primero a :
- generar nuestros datos aleatoriamente
- mostrar los datos generados

```python


#ahora vamos a generar nuestros datos con blobs de sklearn
from matplotlib import pyplot
from pandas import DataFrame
from sklearn.datasets.samples_generator import make_blobs

# Generamos clasificador 2D del dataset
X, y = make_blobs(n_samples=10000, centers=3, n_features=2, random_state=0)

# plotear puntos con su valor de color
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))

colors = {0:'red', 1:'blue', 2:'green'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()

```

```python
print(X.shape)
```

### Division de datos para training y test

En este punto vamos hacer uso de la libreria `train_test_split` para dividir nuestro dataset y poder tener un 70% para el entrenamiento y un 30% para el test, asi mas adelante poder entrenar y hacer nuestro test y tambien comprobar si nuestros resultados son clasificador correctamente.

```python
# Importamos la funcion train_test_split
from sklearn.model_selection import train_test_split

# Dividimos el dataset en train y test
# 70% para el training y 30% para el test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=109) 
```

```python
# mostramos nuestros datos con el tamaño 
# separado correcta mente aqui para el 
# training tanto las caracteristicas  y test
X_train.shape, y_train.shape
```

```python
# mostramos el tamaño de los datos separados para hacer
# la prueba mas adelante con los datos restantes
# aqui, tenemos el test
X_test.shape, y_test.shape
```

### Training y Predicción con SVM (sklearn)

En esta parte hacemos uso de `svm`, importantdo desde la libreria sklearn de tal manera que acontinuación mostramos 
como este actua cuando probamos un cambio en sus caracteristicas y tambien podemos ver como actua cuando :
- se cancbiam los kernesl,lineal, polinomial, sigmoidal y gaussiano.
- tambien podemos ver como influye el parametro c dentro de nuestro modelo.

```python
# Importamos el modelo SVM de sklearn
from sklearn import svm

#usar el clasificador SVM
clf = svm.SVC(C=1.0, kernel='linear') # kernel Lineal

# Modelo de Train usando el conjunto de entrenamiento
clf.fit(X_train, y_train)

# Prediccion del modelo del conjunto de test
y_pred = clf.predict(X_test)
```

```python
#Importamos el modulo metrics para calcular el accuracy
from sklearn import metrics
# experimentaremos con parametros c={0.1, 0.2.0.3..0.9}
ejeX = []
ejeY = []
aux_c = 0.00001;
for i in range(0, 10):    
    #print(aux_c)
    clf = svm.SVC(aux_c, kernel='linear') # kernel Lineal
    clf.fit(X_train, y_train)
    # Prediccion del modelo del conjunto de test
    y_pred = clf.predict(X_test)
    # Modelo Accuracy
    ejeX.append(aux_c)
    ejeY.append(metrics.accuracy_score(y_test, y_pred))
    print("%.5f"%aux_c," Acc:%.3f"%metrics.accuracy_score(y_test, y_pred))    
    aux_c+= 0.00001
```

```python
pyplot.plot(ejeX,ejeY)
pyplot.xlabel("Parametro C")
pyplot.ylabel("Accuracy")
```

 aqui finalmente mostramos la medida de accuracy obtenido con svm para el modelo lineal con un parametro c = 0.1

```python


# Modelo Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
